# Fit polynomial - tension de la pile

```python
import numpy as np
import matplotlib.pyplot as plt

# x : valeurs d'entrée sur alimentation
x_values = np.array([1.6, 1.7, 1.8, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5])  
# y : valeurs mesurées à l'entrée de l'ADC 
y_values = np.array([1.1, 1.18, 1.23, 1.33, 1.5, 1.62, 1.7, 1.77, 1.83]) 


# Fit polynomial: x = f(y)
coeffs = np.polyfit(y_values, x_values, deg=2)
poly_func = np.poly1d(coeffs)

plt.scatter(y_values, x_values, label="Données expérimentales", marker='o')

y_fit = np.linspace(min(y_values), max(y_values), 300)
x_fit = poly_func(y_fit)
plt.plot(y_fit, x_fit, label="Fit polynomial (x = f(y))")

plt.xlabel("Valeur mesurée (y) [V]")
plt.ylabel("Valeur réelle (x) [V]")
plt.title("Correction de mesure : x = f(y)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print("Coefficients du polynôme (x = a*y² + b*y + c):")
print(f"a = {coeffs[0]:.6f}, b = {coeffs[1]:.6f}, c = {coeffs[2]:.6f}")
```
